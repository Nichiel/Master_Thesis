```python
import pandas as pd
from ranking_ut import *
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('Studencka_up.csv')
data_dict = data.to_dict('list')
keys = list(data_dict.keys())

animals_dict = return_animals_dict()
metric_ranking =  ranking_ut(data, animals_dict)
```

```python
fig, axs = plt.subplots(5, 2, figsize=(6, 15), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()

for i, (key, value) in enumerate(metric_ranking.items()):
    axs[i].bar(range(len(value)), list(value.values()), align='center', color=['tab:orange', 'tab:green', 'tab:blue'])
    axs[i].bar_label(axs[i].containers[0])
    axs[i].set_xticks(range(len(value)))
    axs[i].set_xticklabels(list(value.keys()))
    axs[i].set_ylabel('Ilo punktw')
    axs[i].set_xlabel('Metoda')
    axs[i].set_ylim([0,200])
    axs[i].set_title(key)


plt.show()
```

```python
# make diagram for ACE RISE and Protopnet
import matplotlib.pyplot as plt
import numpy as np


#plot bar high for ech method
methods_ranking, ranking_value, percent_ranking_value = return_methods_ranking(metric_ranking)
plt.figure()
plt.bar(range(len(methods_ranking)), list(methods_ranking.values()), align='center', color=['tab:orange', 'tab:green', 'tab:blue'])
plt.bar_label(plt.bar(range(len(methods_ranking)), list(methods_ranking.values()), align='center', color=['tab:orange', 'tab:green', 'tab:blue']), padding=3)
plt.xticks(range(len(methods_ranking)), list(methods_ranking.keys()))
plt.title("Ranking metod")
plt.ylabel("Liczba punktw")
plt.xlabel("Metoda")
plt.ylim([0,1200])
plt.show()
sum(methods_ranking.values())
```

```python
methods_ranking = {"ACE": 0, "RISE": 0, "Protopnet": 0}
#print the best method for each animal
for key, value in metric_ranking.items():
    methods_ranking[max(value, key=value.get)] += 1

#make winner diagram
plt.figure()
plt.bar(range(len(methods_ranking)), list(methods_ranking.values()), align='center', color=['tab:orange', 'tab:green', 'tab:blue'])
plt.xticks(range(len(methods_ranking)), list(methods_ranking.keys()))
plt.title("Ranking zwycistw")
plt.ylabel("Liczba zwycistw")
plt.xlabel("Metoda")
plt.show()
```
